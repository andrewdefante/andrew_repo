```python
import fastf1 as ff1
import pandas as pd
from fastf1 import plotting
from matplotlib import pyplot as plt
```

```python
# load all sessions for 2021 grand prix
bahrain_21 = ff1.get_session(2021, 'Bahrain', 'FP1')
bahrain_21 = bahrain_21.load_laps(with_telemetry=True)
bahrain_21 = pd.DataFrame(data = bahrain_21)
bahrain_21.to_csv('bahrain_21_p.csv')
```

```python
imola_21 = ff1.get_session(2021, 'Imola', 'R')
imola_21 = imola_21.load_laps(with_telemetry=True)
imola_21 = pd.DataFrame(imola_21)
imola_21.to_csv('imola_21.csv')
```

```python
# return bahrain gp quali times from 2020
bahrain_q_20 = ff1.get_session(2020, 'Bahrain', 'Q').load_laps()

# return bahrain gp quali times from 2021
bahrain_q_21 = ff1.get_session(2021, 'Bahrain', 'Q').load_laps()
```

```python
bahrain_q_20 = pd.DataFrame(data = bahrain_q_20)
```

```python
bahrain_q_21 = pd.DataFrame(data = bahrain_q_21)
bahrain_q_21.describe()
```

```python
# return bahrain gp quali times from 2021
bahrain_r_21 = ff1.get_session(2021, 'Bahrain', 'R').load_laps()
```

```python
bahrain_gp_21 = pd.DataFrame(data = bahrain_r_21)
```

```python
bahrain_gp_21.to_csv('bahrain_r_21.csv')
```

```python
# verstappen race pace bahrain gp 2021

bhr21_ver = bahrain_gp_21[bahrain_gp_21["Driver"]== 'VER']
bhr21_ver = bhr21_ver[bhr21_ver["TrackStatus"]=='1']
bhr21_ver.head()
```

```python
import plotly.express as px
```

```python
fig = px.line(bhr21_ver, x="TyreLife", y='LapTime', color = "Stint")
fig
```

```python
# define variables 
tire = bhr21_ver['TyreLife']
lt = bhr21_ver['LapTime']

# The rest is just plotting
fig, ax = plt.subplots()
ax.plot(tire, lt, label='Fast')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.set_title('Leclerc is')
ax.legend()
plt.show()
```
